```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
facility_file = "current-healthcare-facility-listing.csv"
facility_df = pd.read_csv(facility_file)
facility_df.head()
```

```python
hospital_file = "hosp_la_df.csv"
hospital_df = pd.read_csv(hospital_file)
hospital_df
```

```python
facility_cols = ["OSHPD_ID","FACILITY_NAME","DBA_ADDRESS1", "DBA_CITY", "DBA_ZIP_CODE","COUNTY_NAME", "ER_SERVICE_LEVEL_DESC","LICENSE_TYPE_DESC","LICENSE_CATEGORY_DESC", "LATITUDE","LONGITUDE"]
facility_transformed = facility_df[facility_cols].copy()

facility_transformed = facility_transformed.rename(columns={"OSHPD_ID":"ID", "FACILITY_NAME":"Name","DBA_ADDRESS1":"Address", "DBA_CITY":"City","DBA_ZIP_CODE":"Zip","COUNTY_NAME":"County", "ER_SERVICE_LEVEL_DESC":"ER Service Level", "LICENSE_TYPE_DESC":"License Type","LICENSE_CATEGORY_DESC":"Discription","LATITUDE":"Lat", "LONGITUDE":"Long"})
```

```python
hospital_cols = ["FAC_NO"]
hospital_transformed = hospital_df[hospital_cols].copy()

hospital_transformed = hospital_transformed.rename(columns={"FAC_NO":"ID"})
```

```python
hospital_data = pd.merge(facility_transformed,hospital_transformed, how="inner", on=["ID","ID"])
```

```python
hospital_data
```

```python
hospital_data.to_json("JSON outputs/hospitals.json",orient='records')
```
